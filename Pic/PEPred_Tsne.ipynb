```python

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
%matplotlib inline
def dealdata(tsne_df,a,x,name,name1):    
    ax2=plt.subplot(a)
    p1=tsne_df[(tsne_df.label=="Positive") ]
    p2=tsne_df[(tsne_df.label=="Negative") ]
    x1=p1.values[:,0].real
    y1=p1.values[:,1].real
    x2=p2.values[:,0]
    y2=p2.values[:,1]
    ##plt.plot(x, y1)
    plt.plot(x1, y1,'o', color='#EE7621',label = 'Positive',markersize='4')
    #画曲线2
    plt.plot(x2, y2,'o', color='#36648B',label = 'Negative',markersize='4')
    # ax.scatter(x1,y1,c = 'r',marker = 'o')  
    # ax.scatter(x2,y2,c = 'b',marker = 'o')  
    plt.xlabel('Dimension1',fontsize=9)
    plt.ylabel('Dimension2',fontsize=9)
    plt.title(name1+" "+name,fontsize=12)
    plt.legend(loc="upper right")
    plt.text(-7,5,x,fontsize='13')
    plt.xlim((-5.5, 5.5))
    plt.ylim((-5.5, 5.5))


def dealdataAB(tsne_df,a,x,name,name1):    
    ax2=plt.subplot(a)
    p1=tsne_df[(tsne_df.label=="Positive") ]
    p2=tsne_df[(tsne_df.label=="Negative") ]
    x1=p1.values[:,0].real
    y1=p1.values[:,1].real
    x2=p2.values[:,0]
    y2=p2.values[:,1]
    ##plt.plot(x, y1)
    plt.plot(x1, y1,'o', color='#EE7621',label = 'Positive',markersize='4')
    #画曲线2
    plt.plot(x2, y2,'o', color='#36648B',label = 'Negative',markersize='4')
    # ax.scatter(x1,y1,c = 'r',marker = 'o')  
    # ax.scatter(x2,y2,c = 'b',marker = 'o')  
    plt.xlabel('Dimension1',fontsize=9)
    plt.ylabel('Dimension2',fontsize=9)
    plt.title(name1+" "+name,fontsize=12)
    plt.legend(loc="upper right")
    plt.text(-6,5,x,fontsize='13')
    plt.xlim((-4.5, 6.5))
    plt.ylim((-5.5, 5.5))
    
def dealdata1(tsne_df,a,x,name,name1):    
    ax2=plt.subplot(8,3,a)
    p1=tsne_df[(tsne_df.label=="Positive") ]
    p2=tsne_df[(tsne_df.label=="Negative") ]
    x1=p1.values[:,0].real
    y1=p1.values[:,1].real
    x2=p2.values[:,0]
    y2=p2.values[:,1]
    ##plt.plot(x, y1)
    plt.plot(x1, y1,'o', color='#EE7621',label = 'Positive',markersize='4')
    #画曲线2
    plt.plot(x2, y2,'o', color='#36648B',label = 'Negative',markersize='4')
    # ax.scatter(x1,y1,c = 'r',marker = 'o')  
    # ax.scatter(x2,y2,c = 'b',marker = 'o')  
    plt.xlabel('Dimension1',fontsize=9)
    plt.ylabel('Dimension2',fontsize=9)
    plt.title(name1+" "+name,fontsize=12)
    plt.legend(loc="upper right")
    plt.text(-7,5,x,fontsize='13')
    plt.xlim((-5.5, 5.5))
    plt.ylim((-5.5, 5.5))
    
def dealdata1CP(tsne_df,a,x,name,name1):    
    ax2=plt.subplot(8,3,a)
    p1=tsne_df[(tsne_df.label=="Positive") ]
    p2=tsne_df[(tsne_df.label=="Negative") ]
    x1=p1.values[:,0].real
    y1=p1.values[:,1].real
    x2=p2.values[:,0]
    y2=p2.values[:,1]
    ##plt.plot(x, y1)
    plt.plot(x1, y1,'o', color='#EE7621',label = 'Positive',markersize='4')
    #画曲线2
    plt.plot(x2, y2,'o', color='#36648B',label = 'Negative',markersize='4')
    # ax.scatter(x1,y1,c = 'r',marker = 'o')  
    # ax.scatter(x2,y2,c = 'b',marker = 'o')  
    plt.xlabel('Dimension1',fontsize=9)
    plt.ylabel('Dimension2',fontsize=9)
    plt.title(name1+" "+name,fontsize=12)
    plt.legend(loc="upper right")
    plt.text(-6,5,x,fontsize='13')
    plt.xlim((-4.5, 6.5))
    plt.ylim((-5.5, 5.5))

def dealdata1SB(tsne_df,a,x,name,name1):    
    ax2=plt.subplot(8,3,a)
    p1=tsne_df[(tsne_df.label=="Positive") ]
    p2=tsne_df[(tsne_df.label=="Negative") ]
    x1=p1.values[:,0].real
    y1=p1.values[:,1].real
    x2=p2.values[:,0]
    y2=p2.values[:,1]
    ##plt.plot(x, y1)
    plt.plot(x1, y1,'o', color='#EE7621',label = 'Positive',markersize='4')
    #画曲线2
    plt.plot(x2, y2,'o', color='#36648B',label = 'Negative',markersize='4')
    # ax.scatter(x1,y1,c = 'r',marker = 'o')  
    # ax.scatter(x2,y2,c = 'b',marker = 'o')  
    plt.xlabel('Dimension1',fontsize=9)
    plt.ylabel('Dimension2',fontsize=9)
    plt.title(name1+" "+name,fontsize=12)
    plt.legend(loc="upper right")
    plt.text(-3.2,5,x,fontsize='13')
    plt.xlim((-2.5, 2.5))
    plt.ylim((-5.5, 5.5))
    
def loaddata(datafile):
    return np.array(pd.read_csv(datafile,sep="\t",header=-1)).astype(np.float)

def pcadeal(a):  
   
    #FT
    pca1 = PCA(n_components=2)
    datafile1 = "C:\Users\Administrator\Desktop\Peptides_optim2\Y_R//"+a+"//1\FT.txt"
    XMat1 = loaddata(datafile1) 
    principalComponents1 = pca1.fit_transform(XMat1)
    labels1=pd.read_csv(r"C:\Users\Administrator\Desktop\Peptides_optim2\Y_R//"+a+"//1\FT_class.txt",header=None,index_col=None)
    FT1=np.vstack((principalComponents1.T,labels1.T))
    FT=pd.DataFrame(data=FT1.T,columns=("Dimension1","Dimension2","label"))
    #TT
    pca2 = PCA(n_components=2)
    datafile2 = "C:\Users\Administrator\Desktop\Peptides_optim2\Y_R//"+a+"//1\TT.txt"
    XMat2 = loaddata(datafile2) 
    principalComponents2 = pca2.fit_transform(XMat2)
    labels2=pd.read_csv(r"C:\Users\Administrator\Desktop\Peptides_optim2\Y_R//"+a+"//1\TT_class.txt",header=None,index_col=None)
    TT1=np.vstack((principalComponents2.T,labels2.T))
    TT=pd.DataFrame(data=TT1.T,columns=("Dimension1","Dimension2","label"))
     #TTR
    pca3 = PCA(n_components=2)
    datafile3 = "C:\Users\Administrator\Desktop\Peptides_optim2\Y_R//"+a+"//1\TTR.txt"
    XMat3 = loaddata(datafile3) 
    principalComponents3 = pca3.fit_transform(XMat3)
    labels3=pd.read_csv(r"C:\Users\Administrator\Desktop\Peptides_optim2\Y_R//"+a+"//1\TTR_class.txt",header=None,index_col=None)
    TTR1=np.vstack((principalComponents3.T,labels3.T))
    TTR=pd.DataFrame(data=TTR1.T,columns=("Dimension1","Dimension2","label"))

    return    FT, TT, TTR

if __name__ == "__main__":
  
    plt.figure(figsize=(12,28), dpi=300,facecolor=(1, 1, 1))#平铺画布，设置dpi300,注意，发表文章dpi不能低于300
    plt.subplots_adjust(wspace =0.3, hspace =0.35)#调节子图之间的宽度 
    FT, TT, TTR = pcadeal("AAP")
    dealdata(FT,831,"A","Training","AAP")
    dealdata(TT,832,"B","Main Independent","AAP")
    dealdata(TTR,833,"C","Alternative Independent","AAP")
    FT, TT, TTR = pcadeal("ABP")
    dealdata(FT,834,"D","Training","ABP")
    dealdata(TT,835,"E","Main Independent","ABP")
    dealdataAB(TTR,836,"F","Alternative Independent","ABP")
    
    FT, TT, TTR = pcadeal("ACP")
    dealdata(FT,837,"G","Training","ACP")
    dealdata(TT,838,"H","Main Independent","ACP")
    dealdataAB(TTR,839,"I","Alternative Independent","ACP")
    
    FT, TT, TTR = pcadeal("AIP")
    dealdata1(FT,10,"J","Training","AIP")
    dealdata1(TT,11,"K","Main Independent","AIP")
    dealdata1(TTR,12,"L","Alternative Independent","AIP")
    FT, TT, TTR = pcadeal("AVP")
    dealdata1(FT,13,"M","Training","AVP")
    dealdata1(TT,14,"N","Main Independent","AVP")
    dealdata1(TTR,15,"O","Alternative Independent","AVP")
    FT, TT, TTR = pcadeal("CPP")
    dealdata1(FT,16,"P","Training","CPP")
    dealdata1(TT,17,"Q","Main Independent","CPP")
    dealdata1CP(TTR,18,"R","Alternative Independent","CPP")
    FT, TT, TTR = pcadeal("QSP")
    dealdata1(FT,19,"S","Training","QSP")
    dealdata1(TT,20,"T","Main Independent","QSP")
    dealdata1(TTR,21,"U","Alternative Independent","QSP")
    FT, TT, TTR = pcadeal("SBP")
    dealdata1SB(FT,22,"V","Training","SBP")
    dealdata1SB(TT,23,"W","Main Independent","SBP")
    dealdata1SB(TTR,24,"Y","Alternative Independent","SBP")
#     plt.show()
    plt.savefig('C:\Users\Administrator\Desktop\ALL_PEP2.pdf',bbox_inches='tight')
```
